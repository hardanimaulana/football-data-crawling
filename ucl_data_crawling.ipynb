```python
import requests
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re
import pandas as pd

def addMatch(match_id):
    page_link = "https://www.uefa.com/uefachampionsleague/match/" + match_id.strip() + "/statistics/"
    page = Request(page_link, headers={'User-Agent': 'Mozilla'})
    html = urlopen(page).read()

    soup = BeautifulSoup(html, 'html.parser')
    
    print(page_link)
    
    match = soup.find("title").contents[0].split('|')[0].strip()
    team_home = match.split('-')[0]
    team_away = match.split('-')[1]
    
    print(team_home, team_away)
        
    score_home = int(soup.find("div", {"data-bind":"text: homeGoalsScored"}).contents[0])
    score_away = int(soup.find("div", {"data-bind":"text: awayGoalsScored"}).contents[0])
    attempts_home = int(soup.find(class_="total-attempts--value graph-bar--number-value graph-bar--number-value__home-team").contents[0])
    attempts_away = int(soup.find(class_="total-attempts--value graph-bar--number-value graph-bar--number-value__away-team").contents[0])
    attemptsOn_home = int(soup.find("div", {"data-bind":"text: homeAttempsOn"}).contents[0])
    attemptsOn_away = int(soup.find("div", {"data-bind":"text: awayAttempsOn"}).contents[0])
    attemptsOff_home = int(soup.find("div", {"data-bind":"text: homeAttempsOff"}).contents[0])
    attemptsOff_away = int(soup.find("div", {"data-bind":"text: awayAttempsOff"}).contents[0])
    poss_home = int(soup.find("div", {"data-bind":"text: homeBallPossession + '%'"}).contents[0].strip('%'))
    poss_away = int(soup.find("div", {"data-bind":"text: awayBallPossession + '%'"}).contents[0].strip('%'))
    passAcc_home = int(soup.find("span", {"data-bind":"text: homePassesCompletion + '%'"}).contents[0].strip('%'))
    passAcc_away = int(soup.find("span", {"data-bind":"text: awayPassesCompletion + '%'"}).contents[0].strip('%'))
    passComp_home = int(soup.find("div", {"data-bind":"text: _.find(homeTeam.statistics,{type:'PASSES_COMPLETED'}) ? _.find(homeTeam.statistics,{type:'PASSES_COMPLETED'}).value : 0"}).contents[0])
    passComp_away = int(soup.find("div", {"data-bind":"text: _.find(awayTeam.statistics,{type:'PASSES_COMPLETED'}) ? _.find(awayTeam.statistics,{type:'PASSES_COMPLETED'}).value : 0"}).contents[0])
    distRun_home = float(soup.find(class_="distance-run--value graph-bar--number-value graph-bar--number-value__home-team").contents[0].strip(' km'))
    distRun_away = float(soup.find(class_="distance-run--value graph-bar--number-value graph-bar--number-value__away-team").contents[0].strip(' km'))
    ballRec_home = int(soup.find("div", {"data-bind":"text: homeRecoveredBalls"}).contents[0])
    ballRec_away = int(soup.find("div", {"data-bind":"text: awayRecoveredBalls"}).contents[0])
    tackles_home = int(soup.find("div", {"data-bind":"text: _.find(homeTeam.statistics,{type:'TACKLES_COMPLETED'}) ? _.find(homeTeam.statistics,{type:'TACKLES_COMPLETED'}).value : 0"}).contents[0])
    tackles_away = int(soup.find("div", {"data-bind":"text: _.find(awayTeam.statistics,{type:'TACKLES_COMPLETED'}) ? _.find(awayTeam.statistics,{type:'TACKLES_COMPLETED'}).value : 0"}).contents[0])
#    blocks_home = int(soup.find("div", {"data-bind":"text: _.find(homeTeam.statistics,{type:'BLOCKED'}) ? _.find(homeTeam.statistics,{type:'BLOCKED'}).value : 0"}).contents[0])
#    blocks_away = int(soup.find("div", {"data-bind":"text: _.find(awayTeam.statistics,{type:'BLOCKED'}) ? _.find(awayTeam.statistics,{type:'BLOCKED'}).value : 0"}).contents[0])
    clearances_home = int(soup.find("div", {"data-bind":"text: _.find(homeTeam.statistics,{type:'CLEARANCES_COMPLETED'}) ? _.find(homeTeam.statistics,{type:'CLEARANCES_COMPLETED'}).value : 0"}).contents[0])
    clearances_away = int(soup.find("div", {"data-bind":"text: _.find(awayTeam.statistics,{type:'CLEARANCES_COMPLETED'}) ? _.find(awayTeam.statistics,{type:'CLEARANCES_COMPLETED'}).value : 0"}).contents[0])
    yCard_home = int(soup.find("div", {"data-bind":"text: homeYellowCards"}).contents[0])
    yCard_away = int(soup.find("div", {"data-bind":"text: awayYellowCards"}).contents[0])
    rCard_home = soup.find("div", {"data-bind":"text: homeRedCards"})
    if rCard_home == None:
        rCard_home = 0
    else:
        rCard_home = int(rCard_home.contents[0])
    rCard_away = soup.find("div", {"data-bind":"text: awayRedCards"})
    if rCard_away == None:
        rCard_away = 0
    else:
        rCard_away = int(rCard_away.contents[0])
    
    fouls_home = int(soup.find("div", {"data-bind":"text: _.find(homeTeam.statistics,{type:'FOULS_COMMITTED'}) ? _.find(homeTeam.statistics,{type:'FOULS_COMMITTED'}).value : 0"}).contents[0])
    fouls_away = int(soup.find("div", {"data-bind":"text: _.find(awayTeam.statistics,{type:'FOULS_COMMITTED'}) ? _.find(awayTeam.statistics,{type:'FOULS_COMMITTED'}).value : 0"}).contents[0]) 
        
    dict_match = {'team_home': team_home,
                  'team_away': team_away,
                  'score_home': score_home,
                  'score_away': score_away,
                  'attempts_home': attempts_home,
                  'attempts_away': attempts_away,
                  'attemptsOn_home': attemptsOn_home,
                  'attemptsOn_away': attemptsOn_away,
                  'attemptsOff_home': attemptsOff_home,
                  'attemptsOff_away': attemptsOff_away,
                  'poss_home': poss_home,
                  'poss_away': poss_away,
                  'passAcc_home': passAcc_home,
                  'passAcc_away': passAcc_away,
                  'passComp_home': passComp_home,
                  'passComp_away': passComp_away,
                  'distRun_home': distRun_home,
                  'distRun_away': distRun_away,
                  'ballRec_home': ballRec_home,
                  'ballRec_away': ballRec_away,
                  'tackles_home': tackles_home,
                  'tackles_away': tackles_away,
               #   'blocks_home': blocks_home,
               #   'blocks_away': blocks_away,
                  'clearances_home': clearances_home,
                  'clearances_away': clearances_away,
                  'yCard_home': yCard_home,
                  'yCard_away': yCard_away,
                  'rCard_home': rCard_home,
                  'rCard_away': rCard_away,
                  'fouls_home': fouls_home,
                  'fouls_away': fouls_away}
    
    print(dict_match)
    
    return dict_match
```

```python
def statsOpp(df_match, team):
    
    statsOpp = {}
    
    if df_match['team_home'] == team:
        statsOpp = {
                  'score_opp' : df_match['score_away'],
                  'attempts_opp' : df_match['attempts_away'],
                  'attemptsOn_opp' : df_match['attemptsOn_away'],
                  'attemptsOff_opp' : df_match['attemptsOff_away'],
                  'poss_opp' : df_match['poss_away'],
                  'passAcc_opp' : df_match['passAcc_away'],
                  'passComp_opp' : df_match['passComp_away'],
                  'distRun_opp' : df_match['distRun_away'],
                  'ballRec_opp' : df_match['ballRec_away'],
                  'tackles_opp' : df_match['tackles_away'],
                  'clearances_opp' : df_match['clearances_away'],
                  'yCard_opp' : df_match['yCard_away'],
                  'rCard_opp' : df_match['rCard_away'],
                  'fouls_opp' : df_match['fouls_away']
        }
    
    if df_match['team_away'] == team:
        statsOpp = {
                  'score_opp' : df_match['score_home'],
                  'attempts_opp' : df_match['attempts_home'],
                  'attemptsOn_opp' : df_match['attemptsOn_home'],
                  'attemptsOff_opp' : df_match['attemptsOff_home'],
                  'poss_opp' : df_match['poss_home'],
                  'passAcc_opp' : df_match['passAcc_home'],
                  'passComp_opp' : df_match['passComp_home'],
                  'distRun_opp' : df_match['distRun_home'],
                  'ballRec_opp' : df_match['ballRec_home'],
                  'tackles_opp' : df_match['tackles_home'],
                  'clearances_opp' : df_match['clearances_home'],
                  'yCard_opp' : df_match['yCard_home'],
                  'rCard_opp' : df_match['rCard_home'],
                  'fouls_opp' : df_match['fouls_home']
        }
        
    return statsOpp

def statsTeam(df_match, team):
    
    statsTeam = {}
    
    if df_match['team_home'] == team:
        statsTeam = {
                  'score_team' : df_match['score_home'],
                  'attempts_team' : df_match['attempts_home'],
                  'attemptsOn_team' : df_match['attemptsOn_home'],
                  'attemptsOff_team' : df_match['attemptsOff_home'],
                  'poss_team' : df_match['poss_home'],
                  'passAcc_team' : df_match['passAcc_home'],
                  'passComp_team' : df_match['passComp_home'],
                  'distRun_team' : df_match['distRun_home'],
                  'ballRec_team' : df_match['ballRec_home'],
                  'tackles_team' : df_match['tackles_home'],
                  'clearances_team' : df_match['clearances_home'],
                  'yCard_team' : df_match['yCard_home'],
                  'rCard_team' : df_match['rCard_home'],
                  'fouls_team' : df_match['fouls_home']
        }
        
    if df_match['team_away'] == team:
        statsTeam = {
                  'score_team' : df_match['score_away'],
                  'attempts_team' : df_match['attempts_away'],
                  'attemptsOn_team' : df_match['attemptsOn_away'],
                  'attemptsOff_team' : df_match['attemptsOff_away'],
                  'poss_team' : df_match['poss_away'],
                  'passAcc_team' : df_match['passAcc_away'],
                  'passComp_team' : df_match['passComp_away'],
                  'distRun_team' : df_match['distRun_away'],
                  'ballRec_team' : df_match['ballRec_away'],
                  'tackles_team' : df_match['tackles_away'],
                  'clearances_team' : df_match['clearances_away'],
                  'yCard_team' : df_match['yCard_away'],
                  'rCard_team' : df_match['rCard_away'],
                  'fouls_team' : df_match['fouls_away']
        }
        
    return statsTeam
```

```python
stat_matrix_par = []
stat_matrix_opp_par = []
stat_matrix_bay = []
stat_matrix_opp_bay = []

for i in range (df_allmatch.shape[0]):
    df_Team = statsTeam(df_allmatch.iloc[i], 'Paris')
    df_Opp = statsOpp(df_allmatch.iloc[i], 'Paris')
    if df_Team != {}:
        stat_matrix_par.append(list(df_Team.values()))
    if df_Opp != {}:
        stat_matrix_opp_par.append(list(df_Opp.values()))

stat_mean_par = np.mean(stat_matrix_par, axis=0)
    
for i in range (df_allmatch.shape[0]):
    df_Team = statsTeam(df_allmatch.iloc[i], 'Bayern')
    df_Opp = statsOpp(df_allmatch.iloc[i], 'Bayern')
    if df_Team != {}:
        stat_matrix_bay.append(list(df_Team.values()))
    if df_Opp != {}:
        stat_matrix_opp_bay.append(list(df_Opp.values()))

stat_mean_bay = np.mean(stat_matrix_bay, axis=0)


```

```python
from sklearn import linear_model

x_par = stat_matrix_opp_par
y_par = stat_matrix_par
x_bay = stat_matrix_opp_bay
y_bay = stat_matrix_bay

LM_par  = linear_model.LinearRegression().fit(x_par,y_par)
LM_bay  = linear_model.LinearRegression().fit(x_bay,y_bay)

stat_matrix_pred_par = np.array(stat_mean_par)
stat_matrix_pred_bay = np.array(stat_mean_bay)

#stat_matrix_pred_par.reshape([-1, 2])
#stat_matrix_pred_par.reshape([-1, 2])

#print(stat_matrix_pred)

final_pred_par = LM_par.predict([stat_matrix_pred_bay])
final_pred_bay = LM_par.predict([stat_matrix_pred_par])

final_pred = np.concatenate((final_pred_par, final_pred_bay))
```

```python
def normalize_stat(x, minx, maxx, newminx, newmaxx):
    normalized = newminx + ((x-minx)/(maxx-minx)*(newmaxx-newminx))
    
    if np.isnan(normalized):
        return 0
    
    else:
        return int(normalized)

for par in range(final_pred.shape[1]):

    print("par", par)
    
    if par == 0:
        min_stats = -3
        max_stats = 3
    
    else:
    
        min_stats = np.min(final_pred[:,par])
        max_stats = np.max(final_pred[:,par])
        
        min_stats = np.floor(np.max((min_stats - (max_stats - min_stats)/2),0))
        max_stats = np.ceil(max_stats + (max_stats - min_stats)/2)

    print("min", min_stats)
    print("max", max_stats)
    
    new_min_stats = np.min(df_allmatch.iloc[:, (par+1)*2:(par+1)*2+2].min(axis=0))

    print("new min", new_min_stats)
    new_max_stats = np.max(df_allmatch.iloc[:, (par+1)*2:(par+1)*2+2].max(axis=0))
    print("new max", new_max_stats)                    
    
    print("par", final_pred[0][par])
    print("bay", final_pred[1][par])
        
    for data in range(final_pred.shape[0]):
        final_pred[data][par] = normalize_stat(final_pred[data][par], min_stats, max_stats, new_min_stats, new_max_stats)
    
max_poss = sum(final_pred[:,4])

final_pred[0,4] = np.ceil(final_pred[0,4]/max_poss*100)
final_pred[1,4] = 100 - final_pred[0,4]    
```
